# Replace

```python
import pandas as pd
var = pd.read_csv("Beast.csv")
var
```

# let suppose we want to replace the serial number [it will include the repeating numbers also {such as we have number = 1 in Avg.Volumn column of our csv file}] [if the repeating numbers are matching the list we provided]

```python
var.replace({"Symbol": '[A-Z]'}, 22, regex = True)
```

# replace using method 

```python
var.replace(1, method="ffill") # bfill is also possible
```

# replace with limit: 

```python
# Here only 2 datas will be changed
var.replace(1, method="ffill", limit=2) 
```

```python
# This will replace the actual data
var.replace(1, method="ffill", limit=2, inplace = True) 
```
